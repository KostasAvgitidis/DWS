```python
# Importing all necessary libraries
import sklearn
from scipy import stats
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn import datasets, metrics
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X, y = datasets.load_diabetes(return_X_y=True) #Load the diabetes Dataset
```

```python
X = X[:, np.newaxis, 2] #We just want only one feature
```

```python
linearRegressionModel = LinearRegression() #Instantiate the model
```

```python
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(X,y,test_size =0.3) #Split the data into train and test
```

```python
linearRegressionModel.fit(x_train, y_train) #Train the model
```

```python
y_predicted = linearRegressionModel.predict(x_test) #predict some values
```

```python
# Print Some metrics
print(stats.spearmanr(y_test, y_predicted))
print(f'MSE({metrics.mean_squared_error(y_test, y_predicted)})')
print(f'R2 Score({metrics.r2_score(y_test, y_predicted)})')
```

```python
# Plot outputs
plt.scatter(x_test, y_test,  color='blue',marker='+')
plt.plot(x_test, y_predicted, color='green', linewidth=3)

plt.xticks(())
plt.yticks(())

plt.show()
```
